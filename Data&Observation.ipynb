## 3. 

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"
try:
    response = requests.get(url, headers=headers, timeout=5)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'lxml')
    main_heading = soup.find('h1', id='firstHeading')
    if main_heading:
        print(f"Main Heading: '{main_heading.text.strip()}'")
    else:
        print("Main heading not found.")
except requests.exceptions.RequestException as e:
    print(f"Error: {e}")
```

## 4.

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"
try:
    response = requests.get(url, headers=headers, timeout=5)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'lxml')
    content_div = soup.find('div', id='mw-content-text')
    first_paragraph = None
    if content_div:
        paragraphs = content_div.find_all('p')
        for p in paragraphs:
            text = p.text.strip()
            if text and not p.find_parent('table', class_='infobox'):
                first_paragraph = text
                break
    if first_paragraph:
        print(f"First Paragraph (truncated): '{first_paragraph[:150]}...'")
except requests.exceptions.RequestException as e:
    print(f"Error: {e}")
```

## 5.

```python
url = "http://quotes.toscrape.com/"
try:
    response = requests.get(url, headers=headers, timeout=5)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'lxml')
    first_quote_div = soup.find('div', class_='quote')
    if first_quote_div:
        quote = first_quote_div.find('span', class_='text').text.strip()
        author = first_quote_div.find('small', class_='author').text.strip()
        print(f"Sample Quote: '{quote}'\nSample Author: {author}")
except Exception as e:
    print(f"An error occurred: {e}")
```

## 7. 

```python
url = "http://quotes.toscrape.com/"
try:
    response = requests.get(url, headers=headers, timeout=10)
    soup = BeautifulSoup(response.text, 'lxml')
    all_quote_divs = soup.find_all('div', class_='quote')
    if all_quote_divs:
        authors_list = [div.find('small', class_='author').text.strip() for div in all_quote_divs]
        print("Authors found on the page:")
        for author in set(authors_list):
            print(f"- {author}")
        print(f"\nTotal unique authors found: {len(set(authors_list))}")
except Exception as e:
    print(f"Error: {e}")
```

## 11. 

```python
url = "http://quotes.toscrape.com/"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')
quotes = [{'Quote': d.find('span', class_='text').text, 'Author': d.find('small', class_='author').text} for d in soup.find_all('div', class_='quote')]
df = pd.DataFrame(quotes)
csv_path = 'web_scraped_quotes.csv'
df.to_csv(csv_path, index=False)
print(f"Saved to: {csv_path}")
```

## 12. 

```python
json_path = 'web_scraped_quotes.json'
with open(json_path, 'w', encoding='utf-8') as f:
    json.dump(quotes, f, indent=4)
print(f"Saved to: {json_path}")
```

## 13.

```python
db_path = 'web_scraped_quotes.db'
conn = sqlite3.connect(db_path)
cursor = conn.cursor()
cursor.execute('CREATE TABLE IF NOT EXISTS quotes (id INTEGER PRIMARY KEY, quote TEXT, author TEXT)')
for q in quotes:
    cursor.execute("INSERT INTO quotes (quote, author) VALUES (?, ?)", (q['Quote'], q['Author']))
conn.commit()
conn.close()
print(f"Saved to: {db_path}")
```

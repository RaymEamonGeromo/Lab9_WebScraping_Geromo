## 3. Send an HTTP GET Request

```python
import requests

# Step 1: Define the target URL [cite: 468]
url = "https://towardsdatascience.com/"

# Step 2: Define a User-Agent header (essential for ethical scraping) [cite: 470, 471]
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}

# Step 3: Send the GET request [cite: 473]
response = requests.get(url, headers=headers)

# Step 4: Verify the server's response (200 is success) [cite: 474, 476]
print("--- Initiating Contact ---")
print(f"Status Code: {response.status_code}")

# Step 5: Preview the raw HTML [cite: 477, 479]
print("HTML Content Preview (First 500 characters):")
print(response.text[:500])
```

## 5. Locate and Extract Specific Data Elements

```python
from urllib.parse import urljoin

url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')

# Target the "In the news" section [cite: 528, 529]
in_the_news_div = soup.find('div', id='mp-itn')

if in_the_news_div:
    print("\n--- From 'In the news' section ---")
    links = in_the_news_div.find_all('a', href=True)
    for link in links[:5]: # Take first 5 [cite: 571]
        title_text = link.text.strip()
        absolute_url = urljoin(url, link['href'])
        print(f"Title: {title_text} | URL: {absolute_url}")
```

## 6. Extract Data from HTML Tables

```python
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_datasets_for_machine_learning_research"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')

# Identify table with class 'wikitable' [cite: 604]
target_table = soup.find('table', class_='wikitable')
if target_table:
    # Step 2: Extract headers [cite: 606]
    table_headers = [th.text.strip() for th in target_table.find('tr').find_all('th')]
    
    # Step 3: Extract data rows [cite: 609, 610]
    table_data = []
    for row in target_table.find_all('tr')[1:6]:
        cells = row.find_all(['td', 'th'])
        table_data.append([cell.text.strip() for cell in cells])
    
    # Step 4: Convert to DataFrame [cite: 616, 620]
    df = pd.DataFrame(table_data, columns=table_headers[:len(table_data[0])])
    print(df.head(3))
```

## 7. Handle Pagination

```python
import time
current_page_url = "http://quotes.toscrape.com/"
all_quotes = []

for page in range(3): # Limit to 3 pages [cite: 644]
    response = requests.get(current_page_url, headers=headers)
    soup = BeautifulSoup(response.text, 'lxml')
    
    # Extract quotes [cite: 651, 655]
    quotes = soup.find_all('div', class_='quote')
    for q in quotes:
        text = q.find('span', class_='text').text
        author = q.find('small', class_='author').text
        all_quotes.append({'text': text, 'author': author})
    
    # Locate Next Page link [cite: 659, 663]
    next_li = soup.find('li', class_='next')
    if next_li:
        next_url = next_li.find('a')['href']
        current_page_url = urljoin("http://quotes.toscrape.com/", next_url)
        time.sleep(1) # Be polite [cite: 670]
    else:
        break
print(f"Total Quotes Collected: {len(all_quotes)}")
```

## 9. Implement Error Handling and Robustness

```python
import requests
from bs4 import BeautifulSoup

test_scenarios = {
    "valid_article_page": "https://en.wikipedia.org/wiki/Artificial_intelligence",
    "non_existent_article": "http://httpbin.org/status/404",
    "slow_server_sim": "http://httpbin.org/delay/6",
    "invalid_domain_connect": "http://this.domain.does.not.resolve.abc"
}

for scenario_name, test_url in test_scenarios.items():
    print(f"\nTesting Scenario: '{scenario_name}' at {test_url}")
    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(test_url, headers=headers, timeout=5)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'lxml')
        main_title_element = soup.find('h1')
        if main_title_element:
            print(f"SUCCESS: '{main_title_element.text.strip()[:100]}'")
    except requests.exceptions.HTTPError as e:
        print(f"ERROR (HTTP): {e.response.status_code}")
    except requests.exceptions.ConnectionError as e:
        print(f"ERROR (Connection): Could not connect to {test_url}")
    except requests.exceptions.Timeout:
        print(f"ERROR (Timeout): Request timed out after 5 seconds")
    except Exception as e:
        print(f"ERROR (Unexpected): {e}")
print("---")
```
